### K Nearest-Neighbors Algorithm

We will test this bare-bones implementation of K nearest-neighbors algorithms on the classic `iris` and `faithful` data sets. Lets start by importing the `pyknn` module and the corresponding data sets.



```python
import pyknn as knn

iris = knn.KNN('iris',n_iter = 6)
faithful = knn.KNN('faithful',n_iter = 6)
```

For simplicity, we will only use K = 2. 

```python
iris.knn(K = 2)
```

```python
faithful.knn(K = 2)
```

It is clear from the two data sets that the convergence of the k nearest-neighbors algorithm depends on the choice of initial mean values. This means that an inappropriate choice of initial values may lead to inaccurate convergence to local minima of the objective function. Additional, comparing the `iris` and `faithful` datasets reveals that differences in the data also change the rate of convergence. Since the `faithful` dataset with more naturally dispersed into two obvious clusters compared to the `iris` dataset, knn has an easier minimizing it's objective function when K = 2.
